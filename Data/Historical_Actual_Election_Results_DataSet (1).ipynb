```python
# Let's combine all the individually merged sets together

import pandas as pd

# define and read back dfs into notebook
df_list = [
    pd.read_csv("1936_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1940_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1944_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1948_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1952_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1956_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1960_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1964_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1968_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1972_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1976_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1980_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1984_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1988_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1992_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("1996_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("2000_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("2004_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("2008_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("2012_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("2016_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("2020_Poll_vs_Actual_Combined.csv"),
    pd.read_csv("2024_Poll_vs_Actual_Combined.csv")
]

# Concatenate into one master df
df_all_years_combined = pd.concat(df_list, ignore_index=True)

display(df_all_years_combined.head(50))

display(df_all_years_combined.columns)
```

```python
# Column diff btw poll leading margin and electoral leading margin
# too see how far away from the actual value the poll results were
df_all_years_combined['Poll_vs_Electoral_Margin_Diff'] = (df_all_years_combined['Poll_Leading_Margin'] - df_all_years_combined['Electoral_Leading_Margin (difference between dem and rep electorial pct)']).round(2)

# Column diff btw poll leading margin and popular leading margin
# too see how far away from the actual value the poll results were
df_all_years_combined['Poll_vs_Popular_Margin_Diff'] = (df_all_years_combined['Poll_Leading_Margin'] - df_all_years_combined['Popular_Leading_Margin (difference between dem and rep popular vote pct)']).round(2)


display(df_all_years_combined.head(50))
```

```python
# Create election_winner column based on who had more electoral votes
df_all_years_combined['election_winner'] = (df_all_years_combined['Republican_Electoral'] > df_all_years_combined['Democrat_Electoral']).astype(int)

display(df_all_years_combined.head(50))
```

```python
df_all_years_combined.columns
```

```python
# Reorder columns
df_all_years_combined = df_all_years_combined[
    ['year', 
     'Democratic', 'Republican', 
     'Poll_Leading_Margin', 'Poll_Leader', 'election_winner',
     'Republican_Electoral', 'Democrat_Electoral', 
     'Republican_Popular', 'Democrat_Popular', 
     'Total_Popular_Vote (Total votes cast in Presidential Election)',
     'Republican_Electoral_pct (out of 270)', 'Democrat_Electoral_pct (out of 270)',
     'Republican_Popular_pct (Out of total votes cast in Presidential Election)', 
     'Democrat_Popular_pct (Out of total votes cast in Presidential Election)',
     'Electoral_Leading_Margin (difference between dem and rep electorial pct)',
     'Popular_Leading_Margin (difference between dem and rep popular vote pct)',
     'Poll_vs_Electoral_Margin_Diff', 'Poll_vs_Popular_Margin_Diff']
]


display(df_all_years_combined.head(50))
```
