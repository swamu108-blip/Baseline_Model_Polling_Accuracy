```python
# Consolidated into one code block to make it easier to scrape data for project. 
# Each print statement should allow me to see the output as it is running, incase any changes or tweaks need to be made. 
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load the Wiki page
url6 = "https://en.wikipedia.org/wiki/Polling_for_United_States_presidential_elections#1956"
response6 = requests.get(url6)
soup6 = BeautifulSoup(response6.content)

# Status of request
print(f'Status Code of Wikipedia site is: {response6.status_code}')

# Get all tables in wikitable class
tables6 = soup6.find_all("table", class_="wikitable")
print(f'Number of tables is: {len(tables6)}')

# Load the 1948 polling table manually by index
df_6 = pd.read_html(str(tables6[5]))[0]

# Show raw table
print('\nRaw Table Preview:')
display(df_6.head(250))

# Rename candidate columns for consistency
df_6 = df_6.rename(columns={
    df_6.columns[2]: 'Democratic',
    df_6.columns[1]: 'Republican'
})

print('\nRenamed Columns Table Preview:')
display(df_6.head(250))

# Show data types before conversion
print('\nData Types Before Conversion:')
display(df_6.dtypes)

# Convert percentage strings to floats
df_6['Democratic'] = df_6['Democratic'].str.replace('%', '').astype(float)
df_6['Republican'] = df_6['Republican'].str.replace('%', '').astype(float)

# Show data types after conversion
print('\nData Types After Conversion:')
display(df_6.dtypes)

# Add calculated columns
df_6['Poll_Leading_Margin'] = abs(df_6['Democratic'] - df_6['Republican'])
df_6['Poll_Leader'] = (df_6['Republican'] > df_6['Democratic']).astype(int)

print('\nTable with New Columns:')
display(df_6.head(250))

print(list(df_6.columns))

# Drop the extra columns
df_6 = df_6.drop(columns='Month')
#df_6 = df_6.drop(columns=['Henry A. Wallace (Progressive)\xa0%', 'Strom Thurmond (Dixiecrat)\xa0%'])

print('\nDropped Extra Columns:')
display(df_6.head(250))

# Drop final summary rows (row 12 and 13)
df_6 = df_6.drop(index=[12, 13])
df_6.reset_index(drop=True, inplace=True)

print('\nAfter Dropping Final Rows (12 & 13):')
display(df_6.tail())

print('\nFinal dtypes:')
display(df_6.dtypes)

# Add year column at the beginning
df_6.insert(0, 'year', 1956)

df_6 = df_6.dropna(subset=['Democratic', 'Republican'], how='all')
df_6 = df_6.reset_index(drop=True)

print('\nFinal Table with Year Column:')
display(df_6.head(250))

print('\nNumber of Nulls left:')
display(df_6.isnull().sum())

df_6.to_csv("1956_Opinion_Polling_Data_cleaned.csv", index=False)
```
