```python
# Consolidated into one code block to make it easier to scrape data for project. 
# Each print statement should allow me to see the output as it is running, incase any changes or tweaks need to be made. 
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load the Wiki page
url5 = "https://en.wikipedia.org/wiki/Polling_for_United_States_presidential_elections#1952"
response5 = requests.get(url5)
soup5 = BeautifulSoup(response5.content)

# Status of request
print(f'Status Code of Wikipedia site is: {response5.status_code}')

# Get all tables in wikitable class
tables5 = soup5.find_all("table", class_="wikitable")
print(f'Number of tables is: {len(tables5)}')

# Load the 1948 polling table manually by index
df_5 = pd.read_html(str(tables5[4]))[0]

# Show raw table
print('\nRaw Table Preview:')
display(df_5.head(250))

# Rename candidate columns for consistency
df_5 = df_5.rename(columns={
    df_5.columns[2]: 'Democratic',
    df_5.columns[1]: 'Republican'
})

print('\nRenamed Columns Table Preview:')
display(df_5.head(250))

# Show data types before conversion
print('\nData Types Before Conversion:')
display(df_5.dtypes)

# Convert percentage strings to floats
df_5['Democratic'] = df_5['Democratic'].str.replace('%', '').astype(float)
df_5['Republican'] = df_5['Republican'].str.replace('%', '').astype(float)

# Show data types after conversion
print('\nData Types After Conversion:')
display(df_5.dtypes)

# Add calculated columns
df_5['Poll_Leading_Margin'] = abs(df_5['Democratic'] - df_5['Republican'])
df_5['Poll_Leader'] = (df_5['Republican'] > df_5['Democratic']).astype(int)

print('\nTable with New Columns:')
display(df_5.head(250))

print(list(df_5.columns))

# Drop the extra columns
df_5 = df_5.drop(columns='Month')
#df_5 = df_5.drop(columns=['Henry A. Wallace (Progressive)\xa0%', 'Strom Thurmond (Dixiecrat)\xa0%'])

print('\nDropped Extra Columns:')
display(df_5.head(250))

# Drop final summary rows (row 10 and 11)
df_5 = df_5.drop(index=[10, 11])
df_5.reset_index(drop=True, inplace=True)

print('\nAfter Dropping Final Rows (10 & 11):')
display(df_5.tail())

print('\nFinal dtypes:')
display(df_5.dtypes)

# Add year column at the beginning
df_5.insert(0, 'year', 1952)

df_5 = df_5.dropna(subset=['Democratic', 'Republican'], how='all')
df_5 = df_5.reset_index(drop=True)

print('\nFinal Table with Year Column:')
display(df_5.head(250))

print('\nNumber of Nulls left:')
display(df_5.isnull().sum())

df_5.to_csv("1952_Opinion_Polling_Data_cleaned.csv", index=False)
```
