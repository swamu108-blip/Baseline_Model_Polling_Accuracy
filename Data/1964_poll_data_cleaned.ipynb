```python
# Consolidated into one code block to make it easier to scrape data for project. 
# Each print statement should allow me to see the output as it is running, incase any changes or tweaks need to be made. 
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load the Wiki page
url8 = "https://en.wikipedia.org/wiki/Polling_for_United_States_presidential_elections#1964"
response8 = requests.get(url8)
soup8 = BeautifulSoup(response8.content)

# Status of request
print(f'Status Code of Wikipedia site is: {response8.status_code}')

# Get all tables in wikitable class
tables8 = soup8.find_all("table", class_="wikitable")
print(f'Number of tables is: {len(tables8)}')

# Load the 1960 polling table manually by index
df_8 = pd.read_html(str(tables8[7]))[0]

# Show raw table
print('\nRaw Table Preview:')
display(df_8.head(250))

# Rename candidate columns for consistency
df_8 = df_8.rename(columns={
    df_8.columns[1]: 'Democratic',
    df_8.columns[2]: 'Republican'
})

print('\nRenamed Columns Table Preview:')
display(df_8.head(250))

# Show data types before conversion
print('\nData Types Before Conversion:')
display(df_8.dtypes)

# Convert percentage strings to floats
df_8['Democratic'] = df_8['Democratic'].str.replace('%', '').astype(float)
df_8['Republican'] = df_8['Republican'].str.replace('%', '').astype(float)

# Show data types after conversion
print('\nData Types After Conversion:')
display(df_8.dtypes)

# Add calculated columns
df_8['Poll_Leading_Margin'] = abs(df_8['Democratic'] - df_8['Republican'])
df_8['Poll_Leader'] = (df_8['Republican'] > df_8['Democratic']).astype(int)

print('\nTable with New Columns:')
display(df_8.head(250))

print(list(df_8.columns))

# Drop the extra columns
df_8 = df_8.drop(columns='Month')
#df_8 = df_8.drop(columns=['Henry A. Wallace (Progressive)\xa0%', 'Strom Thurmond (Dixiecrat)\xa0%'])

print('\nDropped Extra Columns:')
display(df_8.head(250))

# Drop final summary rows (row 9 and 10)
df_8 = df_8.drop(index=[9, 10])
df_8.reset_index(drop=True, inplace=True)

print('\nAfter Dropping Final Rows (9 & 10):')
display(df_8.tail())

print('\nFinal dtypes:')
display(df_8.dtypes)

# Add year column at the beginning
df_8.insert(0, 'year', 1964)

df_8 = df_8.dropna(subset=['Democratic', 'Republican'], how='all')
df_8 = df_8.reset_index(drop=True)

print('\nFinal Table with Year Column:')
display(df_8.head(250))

print('\nNumber of Nulls left:')
display(df_8.isnull().sum())

df_8.to_csv("1964_Opinion_Polling_Data_cleaned.csv", index=False)
```
