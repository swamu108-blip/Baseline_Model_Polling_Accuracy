```python
# Consolidated into one code block to make it easier to scrape data for project. 
# Each print statement should allow me to see the output as it is running, incase any changes or tweaks need to be made. 
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load the Wiki page
url9 = "https://en.wikipedia.org/wiki/Polling_for_United_States_presidential_elections#1968"
response9 = requests.get(url9)
soup9 = BeautifulSoup(response9.content)

# Status of request
print(f'Status Code of Wikipedia site is: {response9.status_code}')

# Get all tables in wikitable class
tables9 = soup9.find_all("table", class_="wikitable")
print(f'Number of tables is: {len(tables9)}')

# Load the 1960 polling table manually by index
df_9 = pd.read_html(str(tables9[8]))[0]

# Show raw table
print('\nRaw Table Preview:')
display(df_9.head(250))

# Rename candidate columns for consistency
df_9 = df_9.rename(columns={
    df_9.columns[2]: 'Democratic',
    df_9.columns[1]: 'Republican'
})

print('\nRenamed Columns Table Preview:')
display(df_9.head(250))

# Show data types before conversion
print('\nData Types Before Conversion:')
display(df_9.dtypes)

# Convert percentage strings to floats
df_9['Democratic'] = df_9['Democratic'].str.replace('%', '').astype(float)
df_9['Republican'] = df_9['Republican'].str.replace('%', '').astype(float)

# Show data types after conversion
print('\nData Types After Conversion:')
display(df_9.dtypes)

# Add calculated columns
df_9['Poll_Leading_Margin'] = abs(df_9['Democratic'] - df_9['Republican'])
df_9['Poll_Leader'] = (df_9['Republican'] > df_9['Democratic']).astype(int)

print('\nTable with New Columns:')
display(df_9.head(250))

print(list(df_9.columns))

# Drop the extra columns
df_9 = df_9.drop(columns='Month')
df_9 = df_9.drop(columns=['George Wallace (American Independent)\xa0%'])

print('\nDropped Extra Columns:')
display(df_9.head(250))

# Drop final summary rows (row 12 and 13)
df_9 = df_9.drop(index=[12, 13])
df_9.reset_index(drop=True, inplace=True)

print('\nAfter Dropping Final Rows (12 & 13):')
display(df_9.tail())

print('\nFinal dtypes:')
display(df_9.dtypes)

# Add year column at the beginning
df_9.insert(0, 'year', 1968)

df_9 = df_9.dropna(subset=['Democratic', 'Republican'], how='all')
df_9 = df_9.reset_index(drop=True)

print('\nFinal Table with Year Column:')
display(df_9.head(250))

print('\nNumber of Nulls left:')
display(df_9.isnull().sum())

df_9.to_csv("1968_Opinion_Polling_Data_cleaned.csv", index=False)
```
