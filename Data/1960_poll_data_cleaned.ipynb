```python
# Consolidated into one code block to make it easier to scrape data for project. 
# Each print statement should allow me to see the output as it is running, incase any changes or tweaks need to be made. 
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load the Wiki page
url7 = "https://en.wikipedia.org/wiki/Polling_for_United_States_presidential_elections#1960"
response7 = requests.get(url7)
soup7 = BeautifulSoup(response7.content)

# Status of request
print(f'Status Code of Wikipedia site is: {response7.status_code}')

# Get all tables in wikitable class
tables7 = soup7.find_all("table", class_="wikitable")
print(f'Number of tables is: {len(tables7)}')

# Load the 1960 polling table manually by index
df_7 = pd.read_html(str(tables7[6]))[0]

# Show raw table
print('\nRaw Table Preview:')
display(df_7.head(250))

# Rename candidate columns for consistency
df_7 = df_7.rename(columns={
    df_7.columns[1]: 'Democratic',
    df_7.columns[2]: 'Republican'
})

print('\nRenamed Columns Table Preview:')
display(df_7.head(250))

# Show data types before conversion
print('\nData Types Before Conversion:')
display(df_7.dtypes)

# Convert percentage strings to floats
df_7['Democratic'] = df_7['Democratic'].str.replace('%', '').astype(float)
df_7['Republican'] = df_7['Republican'].str.replace('%', '').astype(float)

# Show data types after conversion
print('\nData Types After Conversion:')
display(df_7.dtypes)

# Add calculated columns
df_7['Poll_Leading_Margin'] = abs(df_7['Democratic'] - df_7['Republican'])
df_7['Poll_Leader'] = (df_7['Republican'] > df_7['Democratic']).astype(int)

print('\nTable with New Columns:')
display(df_7.head(250))

print(list(df_7.columns))

# Drop the extra columns
df_7 = df_7.drop(columns='Month')
#df_7 = df_7.drop(columns=['Henry A. Wallace (Progressive)\xa0%', 'Strom Thurmond (Dixiecrat)\xa0%'])

print('\nDropped Extra Columns:')
display(df_7.head(250))

# Drop final summary rows (row 14 and 15)
df_7 = df_7.drop(index=[14, 15])
df_7.reset_index(drop=True, inplace=True)

print('\nAfter Dropping Final Rows (14 & 15):')
display(df_7.tail())

print('\nFinal dtypes:')
display(df_7.dtypes)

# Add year column at the beginning
df_7.insert(0, 'year', 1960)

df_7 = df_7.dropna(subset=['Democratic', 'Republican'], how='all')
df_7 = df_7.reset_index(drop=True)

print('\nFinal Table with Year Column:')
display(df_7.head(250))

print('\nNumber of Nulls left:')
display(df_7.isnull().sum())

df_7.to_csv("1960_Opinion_Polling_Data_cleaned.csv", index=False)
```
